```python
# in the below code as same as above. in this we create a use define funtion, the only difference. 

def time_series(data, x, y, ID):
    Stock_1 = data[data["StockCode"] == ID]

    plt.figure(figsize=(18,7))
    plt.plot(Stock_1[x],Stock_1[y],"o-",markersize=5)
    plt.xlabel("Timestamp",fontsize=13)
    plt.ylabel(y,fontsize=13)
    plt.title(y+ "trend of product " +ID, fontsize=13)
    plt.grid()
```

```python
# using the above function
time_series(data=df_11, x="InvoiceDate", y="Quantity", ID="POST")
```
